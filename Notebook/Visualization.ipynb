```python
import torch
import matplotlib.pyplot as plt
```

# Vanilla_GAN-Linear(Adam optimizer)

```python
loss_d = torch.load("/home/careinfolab/Dr_Luo/Rohan/Gen_Adver_Nets/Results/Vanilla_GAN_Linear/train_loss_d.pth")
loss_g = torch.load("/home/careinfolab/Dr_Luo/Rohan/Gen_Adver_Nets/Results/Vanilla_GAN_Linear/train_loss_g.pth")

plt.plot(loss_d, label="Discriminator Loss")
plt.plot(loss_g, label="Generator Loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.title("Vanilla_GAN Linear Training Loss")
plt.show()
```
